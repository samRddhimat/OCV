### <code style="background:red;color:white">cv2.split()</code>

```python
yara_bgr = cv2.imread("D:/Repository/OCV/yarravalley.jpg",cv2.IMREAD_COLOR) #cv2.IMREAD_COLOR_BGR

# print(type(lake_bgr),cv2.split(yara_bgr))
print(f"type of lake is {type(yara_bgr)}, length is {len(yara_bgr)}, size is {yara_bgr.size},  shape is ",{yara_bgr.shape})
b,g,r = cv2.split(yara_bgr)
```

## <code style="background:red;color:white">BGR to RGB</code>

```python

# img_yarra_valley = cv2.imread("D:/Repository/OCV/yarravalley.jpg",cv2.IMREAD_COLOR)

yarra_rgb = cv2.cvtColor(yarra_bgr,cv2.COLOR_BGR2RGB)
mplt.imshow(yarra_rgb)
mplt.show()
```

## <code style="background:red;color:white">BGR to HSV</code>

<ol>
<li>h - represents the actual color of the image</li>
<li>s - intensity of the color (kinda pure red vs dark red) - How strong or dull a color is</li>
<li>Low saturation → looks grayish</li>
<li>High saturation → looks vibrant and rich</li>

<li>v - How light or dark the color is.</li>

<li>Hue helps isolate specific colors regardless of lighting.</li>
<li>Saturation and Value help with intensity and brightness filtering.</li>
<li>Used in:</li>
<li>Skin tone detection</li>
<li>Color tracking</li>
<li>Object segmentation</li>
</ol>

```python
yarra_hsv = cv2.cvtColor(yarra_bgr,cv2.COLOR_BGR2HSV)
h,s,v = cv2.split(yarra_hsv)

# mplt.figure(figsize=[15,15])
# mplt.subplot(141);mplt.imshow(h,cmap='gray');mplt.title("H channel");
# mplt.show()

# mplt.figure(figsize=[15,15])
# mplt.subplot(142);mplt.imshow(s,cmap='gray');mplt.title("H channel");
# mplt.show()

# mplt.figure(figsize=[15,15])
# mplt.subplot(143);mplt.imshow(v,cmap='gray');mplt.title("H channel");
# mplt.show()

fighvs,axshvs = mplt.subplots(1,4,figsize=(10,10))

axshvs[0].set_title("Red Channel")
axshvs[0].imshow(h,cmap='gray')
axshvs[0].title.set_size(10)

axshvs[1].set_title("Green Channel")
axshvs[1].imshow(v,cmap='gray')
axshvs[1].title.set_size(10)

axshvs[2].set_title("Blue Channel")
axshvs[2].imshow(s,cmap='gray')
axshvs[2].title.set_size(10)

axshvs[3].set_title("Original")
axshvs[3].imshow(yarra_rgb)
axshvs[3].title.set_size(10)

fighvs.tight_layout()
mplt.show()
```

### <code style="background:red;color:white">Modifying Individual Channel</code>

```python
h_new = h + 10
h.shape
# h[193][100]
```

```python
yarra_merged = cv2.merge((h_new,s,v))
yarra_rgb = cv2.cvtColor(yarra_merged,cv2.COLOR_HSV2RGB)

figmerged,axsmerged = mplt.subplots(1,4,figsize=(10,10))

axsmerged[0].set_title("Red Channel")
axsmerged[0].imshow(h,cmap='gray')
axsmerged[0].title.set_size(10)

axsmerged[1].set_title("Green Channel")
axsmerged[1].imshow(v,cmap='gray')
axsmerged[1].title.set_size(10)

axsmerged[2].set_title("Blue Channel")
axsmerged[2].imshow(s,cmap='gray')
axsmerged[2].title.set_size(10)


axsmerged[3].set_title("Modified")
axsmerged[3].imshow(yarra_rgb)
axsmerged[3].title.set_size(10)
figmerged.tight_layout()
mplt.show()
```
